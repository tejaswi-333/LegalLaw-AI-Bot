```python
import sqlite3
import json

conn = sqlite3.connect("chat_history.db")
cursor = conn.cursor()

cursor.execute("SELECT session_id, messages FROM sessions")
rows = cursor.fetchall()

for session_id, messages in rows:
    print("\nSession ID:", session_id)
    print(json.dumps(json.loads(messages), indent=4))

conn.close()
```

```python
import sqlite3
import json

conn = sqlite3.connect("chat_history.db")
cursor = conn.cursor()

cursor.execute("SELECT id, full_name, email, created_at FROM users")
rows = cursor.fetchall()

users_json = []

for row in rows:
    users_json.append({
        "no": row[0],
        "username": row[1],
        "emailid": row[2],
        "timestamp": row[3]
    })

# Print each user on separate lines (pretty JSON)
for user in users_json:
    print(json.dumps(user, indent=4))

conn.close()
```
